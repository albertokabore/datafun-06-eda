# Iris Dataset Analysis

Our goal is to explore and understand the relationships between these measurements and how they vary across different species.

•	These observations suggest that while there is some variation in sepal size, the petal sizes are quite uniform in the early entries of the dataset.

# Step 3. Initial Descriptive Statistics


```python
print(df.describe())
```

•	Quartiles for each attribute underscore the data's distribution, revealing that median values (50th percentile) for petal length and petal width are notably higher than the first quartile, suggesting a skewed distribution for these features.



# Step 4. Initial Data Distribution for Numerical Columns


```python
# Inspect histogram by numerical column
df['sepal_length'].hist()

# Inspect histograms for all numerical columns
df.hist()

# Show all plots
plt.show()
```

•	These distributions underscore the variability and characteristics of the Iris flower dimensions, potentially reflecting differences between species. Each histogram illustrates the central tendency and spread, which are crucial for statistical analysis and machine learning model training.



# Step 5. Initial Data Distribution for Categorical Columns


```python
# Inspect value counts by categorical column
df['species'].value_counts()

# Inspect value counts for all categorical columns
for col in df.select_dtypes(include=['object', 'category']).columns:
    # Display count plot
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()

# Show all plots
plt.show()
```

Observatio:

•	The bar chart shows an even distribution among the three species of the Iris dataset: setosa, versicolor, and virginica.

•	Each species has approximately the same number of samples, suggesting that the dataset is balanced. This balance is ideal for statistical analysis and machine learning, as it avoids bias toward a particular species.

•	The uniformity in the number of observations per species ensures that any analysis, such as comparative studies on morphological traits or model training, will not be skewed by uneven sample sizes.

•	Such a distribution facilitates reliable comparisons and accurate predictions across the different Iris species in further analyses or modeling.



# Step 6. Initial Data Transformation and Feature Engineering


1. Renaming a Column



```python
df.rename(columns={'species': 'flower_species'}, inplace=True)
```

```python
# Verify transformations
print(df.head())
```

2. Adding New Columns.

```python
# Adding a new column 'sepal_area' as the product of sepal length and sepal width
df['sepal_area'] = df['sepal_length'] * df['sepal_width']
```

```python
# Verify transformations
print(df.head())
```

```python
# Display the updated DataFrame with the new column
print(df.head(10))
```

# Step 7. Initial Visualizations

1. Distribution of Sepal Lengths by Species


Goal:
Explore the distribution of sepal lengths across different species to understand if there are distinct differences

Chart Type:
Violin Plot

```python
# Violin plot to show distribution of sepal lengths by species
plt.figure(figsize=(10, 6))
sns.violinplot(x='flower_species', y='sepal_length', data=df)
plt.title('Distribution of Sepal Lengths by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

Observation:

The violin plot shows the distribution of sepal lengths for each species in the Iris dataset.
Setosa:

•	Narrow distribution centered around shorter sepal lengths (~5 cm).

•	Indicates less variability in sepal length.
Versicolor:

•	Broader distribution with sepal lengths primarily ranging between 5 and 7 cm.

•	Indicates moderate variability in sepal length.

Virginica:

•	Widest distribution with sepal lengths spanning from 5.5 to 8 cm.

•	Indicates high variability in sepal length.

Sepal length is a distinguishing feature among the species, with Setosa having distinctly shorter sepal lengths on average compared to Versicolor and Virginica.



2. Distribution of Sepal Lengths

Goal:
Explore the distribution of sepal lengths across different species.

Chart Type:
Histogram

```python
# Histogram for sepal length
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='sepal_length', hue='flower_species', multiple='stack', palette='viridis')
plt.title('Distribution of Sepal Lengths by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Count')
plt.show()
```

Observations:

The histogram displays the distribution of sepal lengths for each species in the Iris dataset.
Setosa:

•	Primarily concentrated between 4.5 and 5.5 cm.

•	Majority of Setosa sepal lengths are less than 6 cm.
Versicolor:

•	Spans from 5.0 to 7.0 cm.

•	Overlaps significantly with both Setosa and Virginica but generally falls between Setosa and Virginica in terms of sepal length.
Virginica:

•	Ranges from 5.5 to 8.0 cm.

•	Predominantly found above 6 cm, indicating generally longer sepal lengths.

The histogram highlights that Setosa has the shortest sepal lengths, Virginica has the longest, and Versicolor overlaps with both but mostly occupies the middle range.



4. Relationship Between Sepal Length and Sepal Width

Goal:
Examine the relationship between sepal length and sepal width across different species.

Chart Type:
Scatter Plot

```python
# Scatter plot for sepal length vs. sepal width
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='flower_species', palette='viridis')
plt.title('Sepal Length vs. Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

Observations:

The scatter plot shows the relationship between sepal length and sepal width for each species in the Iris dataset.

Setosa:

•	Clusters distinctly with shorter sepal lengths (4.3 to 5.8 cm) and wider sepal widths (3.0 to 4.4 cm).

•	Forms a separate group from the other two species.

Versicolor:

•	Sepal lengths range from 4.9 to 7.0 cm with sepal widths between 2.0 and 3.4 cm.

•	Overlaps partially with Virginica, showing more variability.

Virginica:

•	Sepal lengths span from 4.9 to 7.9 cm and sepal widths from 2.2 to 3.8 cm.

•	Generally, has longer sepal lengths and overlaps with Versicolor but is still distinguishable.

The plot highlights clear separation between Setosa and the other species, while Versicolor and Virginica show some overlap but differ in average sepal dimensions.



5. Sepal Area Distribution

Goal:
Investigate the distribution of the newly created sepal area feature across different species.

Chart Type:
Box plot

```python
# Box plot for sepal area
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='flower_species', y='sepal_area', hue='flower_species', palette='viridis', dodge=False)
plt.title('Distribution of Sepal Area by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Area (cm^2)')
plt.legend([],[], frameon=False)
plt.show()
```

Observation:

The box plot illustrates the distribution of sepal area across different species in the Iris dataset.
Setosa:

•	Median sepal area around 17.5 cm².

•	Interquartile range (IQR) extends from approximately 15 cm² to 20 cm².

•	No outliers observed.

Versicolor:

•	Median sepal area slightly lower than Setosa, around 16.5 cm².

•	IQR extends from about 15 cm² to 20 cm².

•	No significant outliers observed.

Virginica:

•	Highest median sepal area, around 20 cm².

•	IQR extends from approximately 17 cm² to 23 cm².

•	A few outliers above 27 cm².

The plot indicates that Virginica has a larger sepal area compared to Setosa and Versicolor, with more variability and a few outliers.
