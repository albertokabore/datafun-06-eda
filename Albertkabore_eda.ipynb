# Step 7. Initial Visualizations

Subsection 1: Exploring the Relationship Between Flipper Length and Body Mass


Goal: 
Investigate how flipper length correlates with body mass across different penguin species.



Chart Type: Scatter Plot

```python
# Scatter plot of flipper length vs. body mass
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=df, x="flipper_length_mm", y="body_mass_g", hue="Penguin Species")
scatter_plot.set_xlabel("Flipper Length (mm)")
scatter_plot.set_ylabel("Body Mass (g)")
scatter_plot.set_title("Penguin Flipper Length vs. Body Mass")
plt.show()
```

Observation:

The scatter plot above illustrates the relationship between flipper length and body mass across different penguin species. We can observe distinct clusters for each species, suggesting a correlation between these two features. For instance, the Gentoo species generally has larger body mass and longer flippers compared to Adelie and Chinstrap. This visual insight indicates that flipper length could be a good predictor of body mass and potentially useful for species classification.

Positive Correlation:

•	There is a positive correlation between flipper length and body mass for all penguin species.

Species Clustering:

•	Gentoo penguins generally have longer flippers and higher body mass compared to Adelie and Chinstrap penguins.

•	Adelie and Chinstrap penguins show more overlap but still form distinct clusters.

Range of Values:

•	Flipper lengths range from 170 mm to 230 mm.

•	Body mass ranges from 3000 g to over 6000 g.

Gentoo Penguins:

•	Gentoo penguins exhibit a wider range of body mass and flipper lengths, often exceeding 5000 g in body mass and 220 mm in flipper length.

This scatter plot effectively illustrates the relationship between flipper length and body mass, highlighting species-specific characteristics and clustering.



Chart Type: Regression Plot



```python
# Regression plot of flipper length vs. body mass
plt.figure(figsize=(10, 6))
reg_plot = sns.lmplot(data=df, x="flipper_length_mm", y="body_mass_g", hue="Penguin Species", height=6, aspect=1.5)
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")
plt.title("Regression of Flipper Length vs. Body Mass by Penguin Species")
plt.show()
```

Observation:

The regression plot displays the linear relationship between flipper length and body mass across different penguin species. We see that all three species show a positive correlation, with Gentoo penguins generally having higher body mass and longer flippers. This plot helps in understanding how these two variables are related and emphasizes the differences among species.

Subsection 2: Comparing Bill Length Across Penguin Species


Goal: Compare the distribution of bill lengths among the different penguin species.

Chart Type: Box Plot

```python
# Box plot for bill length
plt.figure(figsize=(10, 6))
box_plot = sns.boxplot(data=df, x="Penguin Species", y="bill_length_mm", hue="Penguin Species", palette="viridis", dodge=False)
box_plot.set_xlabel("Penguin Species")
box_plot.set_ylabel("Bill Length (mm)")
box_plot.set_title("Distribution of Bill Length by Penguin Species")
plt.legend([],[], frameon=False)  # Removes the legend since we are only using hue for color
plt.show()
```

Observation:

The box plot shows the distribution of bill lengths for the three penguin species. We can see that the Gentoo species tends to have longer bills compared to Adelie and Chinstrap. The median bill length is highest for Gentoo, followed by Chinstrap, and then Adelie. The variability in bill length is also evident, with Gentoo displaying the widest range. This visualization helps us understand the physical differences between species, particularly in bill length.

Subsection 5: Correlation Matrix of Penguin Features


Goal: Examine the correlations between different numerical features in the dataset.



Chart Type: Heatmap



```python
# Compute the correlation matrix for numerical columns
numerical_df = df.select_dtypes(include=['float64', 'int64'])
corr = numerical_df.corr()

# Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)
heatmap.set_title("Correlation Matrix of Penguin Features")
plt.show()
```

Observation:

Positive Correlations:

•	Flipper Length and Body Mass: There is a strong positive correlation (0.87) between flipper length and body mass, indicating that penguins with longer flippers tend to have higher body mass.

•	Bill Length and Flipper Length: Bill length and flipper length also show a significant positive correlation (0.66).

Negative Correlations:

•	Bill Depth and Flipper Length: A moderate negative correlation (-0.58) exists between bill depth and flipper length, suggesting that penguins with deeper bills tend to have shorter flippers.

Bill Length and Body Mass:

•	Moderate positive correlations are observed between bill length and body mass (0.6), indicating that penguins with longer bills generally have higher body mass.

Bill Depth:

•	Bill depth shows a moderate negative correlation with both flipper length (-0.58) and body mass (-0.47), indicating an inverse relationship with these features.

Body Mass in Pounds:

•	The correlation matrix confirms that body mass in pounds has a perfect correlation (1.0) with body mass in grams, as expected from the linear conversion.

These observations highlight key relationships between different morphological features of penguins, which can be useful for further analysis and predictive modeling.



Subsection 6: Bill Length and Depth Distribution


Goal: To explore the distribution of bill length and bill depth across different penguin species.



Chart Type: Joint Plot

```python
# Joint plot for bill length vs. bill depth
plt.figure(figsize=(10, 6))
joint_plot = sns.jointplot(data=df, x="bill_length_mm", y="bill_depth_mm", hue="Penguin Species", kind="kde", height=8)
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.suptitle("Joint Distribution of Bill Length and Depth by Penguin Species", y=1.02)
plt.show()
```

Observation:
The count plot shows the distribution of different penguin species across various islands. We can observe that the Adelie species is present on all three islands, while Chinstrap penguins are found only on Dream island. Gentoo penguins are predominantly on Biscoe island. This visualization helps in understanding the geographical distribution of penguin species and can be useful for ecological studies.
